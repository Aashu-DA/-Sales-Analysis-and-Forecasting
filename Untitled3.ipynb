```python
import pandas as pd

#'Amazon_Sales_data,csv' is my file
df = pd.read_csv('/content/drive/My Drive/Amazon_Sales_data.csv')

# Convert the 'Order Date' and 'Ship Date' columns to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
# Extracting month, year, and year-month
df['Order Month'] = df['Order Date'].dt.month
df['Order Year'] = df['Order Date'].dt.year
df['Order Year-Month'] = df['Order Date'].dt.to_period('M')
```

```python
# Calculate total sales
monthly_sales = df.groupby('Order Month')['Total Revenue'].sum()
yearly_sales = df.groupby('Order Year')['Total Revenue'].sum()
year_month_sales = df.groupby('Order Year-Month')['Total Revenue'].sum()
```

```python
import matplotlib.pyplot as plt

# Plotting monthly sales
plt.figure(figsize=(12, 6))
monthly_sales.plot(kind='bar', color='skyblue')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.show()
```

```python
#Average Order Value
avg_order_value = df.groupby('Order Year')['Total Revenue'].mean()
```

```python
# Calculate total sales for each year
yearly_sales = df.groupby('Order Year')['Total Revenue'].sum()

# Visualize yearly sales trend
plt.figure(figsize=(12, 6))
yearly_sales.plot(kind='bar', color='green')
plt.title('Yearly Sales Trend')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Calculate total sales for each year-month
year_month_sales = df.groupby('Order Year-Month')['Total Revenue'].sum()

# Visualize year-month sales trend
plt.figure(figsize=(12, 6))
year_month_sales.plot(kind='line', marker='o', color='orange')
plt.title('Year-Monthly Sales Trend')
plt.xlabel('Year-Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
# Sales distribution
import seaborn as sns
plt.figure(figsize=(12, 6))
sns.histplot(df['Total Revenue'], bins=30, kde=True, color='skyblue')
plt.title('Sales Distribution')
plt.xlabel('Total Revenue')
plt.ylabel('Frequency')
plt.show()
```

```python
# Sales by Region
plt.figure(figsize=(12, 6))
sns.barplot(x='Region', y='Total Revenue', data=df, hue='Region', palette='viridis', ci=None, legend=False)
plt.title('Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.show()
```

```python
# Sales by Item Type with a barplot
plt.figure(figsize=(12, 6))
sns.barplot(x='Item Type', y='Total Revenue', data=df, hue='Item Type', palette='muted', ci=None, legend=False)
plt.title('Sales by Item Type')
plt.xlabel('Item Type')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.show()
```

```python
# Sales Channel Comparison
plt.figure(figsize=(10, 6))
sns.boxplot(x='Sales Channel', y='Total Revenue', data=df, palette='pastel')
plt.title('Sales Channel Comparison')
plt.xlabel('Sales Channel')
plt.ylabel('Total Revenue')
plt.show()
```

```python
# Order Priority Analysis
plt.figure(figsize=(8, 6))
sns.countplot(x='Order Priority', data=df, palette='Set2')
plt.title('Order Priority Analysis')
plt.xlabel('Order Priority')
plt.ylabel('Count')
plt.show()
```

```python
# Extracting only numeric columns for correlation analysis
numeric_columns = df.select_dtypes(include=['number']).columns
numeric_df = df[numeric_columns]

# Correlation matrix
correlation_matrix = numeric_df.corr()

# Visualize the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```
